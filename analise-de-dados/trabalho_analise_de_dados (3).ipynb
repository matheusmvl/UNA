# Descrio do projeto


Este dataset contm uma lista de jogos eletrnicos com vendas superiores a 100.000 cpias no perodo de 1985 a 2016.

<h2>Os campos incluem:</h2>

**Name:** o nome do jogo

**Plataform:** plataforma de lanamento dos jogos (ou seja, PS2, Wiim Xbox, etc.)

**Year:** ano de lanamento do jogo

**Genre:** Gnero do jogo

**Publisher:** Empresa que publicou o jogo

**NA_Sales:** Vendas na Amrica do Norte (em milhes)

**EU_Sales:** Vendas na Europa (em milhes)

**JP_Sales:** Vendas no Japo (em milhes)

**Other_Sales:** Vendas no resto do mundo (em milhes)

**Global_Sales:** Total de vendas mundiais.


# Bibliotecas Utilizadas no projeto

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Configurao do dataset

```
dados = pd.read_csv('vgsales.csv')
dados.head()
```

```
dados.drop(labels='Rank', axis=1, inplace=True)

dicionario = {
    'Name' : 'Nome',
    'Platform' : 'Plataforma',
    'Year' : 'Ano',
    'Genre' : 'Genero',
    'Publisher' : 'Publicadora',
    'NA_Sales' : 'Vendas_NA',
    'EU_Sales' : 'Vendas_EU',
    'JP_Sales' : 'Vendas_JP',
    'Other_Sales' : 'Vendas_Outros',
    'Global_Sales' : 'Vendas_Global'
}

dados.rename(columns=dicionario, inplace=True)
```

```
dados.dropna(how="any",inplace = True)
dados.info()
```

```
dados.Ano = dados.Ano.astype(int)
dados.head()
```

# Frequncia por plataforma

```
frequencia = dados.Plataforma.value_counts()
percentual = dados.Plataforma.value_counts(normalize=True) * 100

distribuicao_plataforma = pd.DataFrame({'Frequncia': frequencia, 'Porcentagem (%)':percentual})
distribuicao_plataforma
```

```
plt.figure(figsize=(15, 10))
sns.countplot(x="Plataforma", data=dados, order = dados['Plataforma'].value_counts().index)
plt.xticks(rotation=90)
```

# Frequncia por ano

```
frequencia = dados.Ano.value_counts()
percentual = dados.Ano.value_counts(normalize=True) * 100

distribuicao_ano = pd.DataFrame({'Frequncia':frequencia, 'Porcentagem (%)':percentual,})
distribuicao_ano
```

```
plt.figure(figsize=(15, 10))
sns.countplot(x="Ano", data=dados, order = dados['Ano'].value_counts().index)
plt.xticks(rotation=90)
```

# Frequncia por gnero

```
frequencia = dados.Genero.value_counts()
percentual = dados.Genero.value_counts(normalize=True) * 100

distribuicao_genero = pd.DataFrame({'Frequencia':frequencia, 'Porcentagem (%)':percentual})
distribuicao_genero
```

```
plt.figure(figsize=(15, 10))
sns.countplot(x="Genero", data=dados, order = dados['Genero'].value_counts().index)
plt.xticks(rotation=90)
```

# Frequncia por Publicadora

```
frequencia = dados.Publicadora.value_counts()
percentual = dados.Publicadora.value_counts(normalize=True) * 100

distribuicao_publicadora = pd.DataFrame({'Frequncia':frequencia, 'Porcentagem (%)':percentual})
distribuicao_publicadora.rename_axis('Publicadora', axis='columns', inplace=True)
distribuicao_publicadora[distribuicao_publicadora['Frequncia'] > distribuicao_publicadora['Frequncia'].mean()]
```

```
plt.figure(figsize=(15, 10))
sns.countplot(x="Publicadora", data=dados, order = dados.groupby(by=['Publicadora'])['Ano'].count().sort_values(ascending=False).iloc[:20].index)
plt.xticks(rotation=90)
```

# Entendendo o comportamento do consumidor por regio

Comparando as vendas de cada regio por gnero.

```
comparacao = dados[['Genero', 'Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros']]
comp_mapa = comparacao.groupby(by=['Genero']).sum()
```

```
plt.figure(figsize=(15, 10))
sns.set(font_scale=1)
sns.heatmap(comp_mapa, annot=True, fmt = '.1f')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

# Comparao de vendas por plataforma

```
comp_plataforma = dados[['Plataforma', 'Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros']]
comp_plataforma.head()
```

```
comp_plataforma = comp_plataforma.groupby(by=['Plataforma']).sum().reset_index()
comp_plataforma['Total'] = comp_plataforma.sum(axis=1)
comp_plataforma_ordenado = comp_plataforma.sort_values(by='Total', ascending=False)
comp_plataforma_ordenado
```

```
plt.figure(figsize=(30, 15))
sns.barplot(x='Plataforma', y='Total', data=comp_plataforma_ordenado)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

# Comparao de vendas por regio

```
top_vendas_regiao = dados[['Vendas_NA', 'Vendas_EU', 'Vendas_JP', 'Vendas_Outros']]
top_vendas_regiao = top_vendas_regiao.sum().reset_index()
top_vendas_regiao = top_vendas_regiao.rename(columns={"index": "Regiao", 0: "Vendas"})
top_vendas_regiao
```

```
labels = top_vendas_regiao['Regiao']
sizes = top_vendas_regiao['Vendas']
```

```
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%0.1f%%', startangle=90)
```
